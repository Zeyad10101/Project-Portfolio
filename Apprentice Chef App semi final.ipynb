```python
# Question NO 1 #
## REGRESSION ANALYSIS ##
#### In order to explore our data and to find out what is the most variables that have negative impact on our revenue, we had to run regression for several variables to test and re-filtered them to the most negative ones by removing those that has positive impact explained by the P value Coefficiency.
#### With the second regression where we eleminated the variable that has postive impact on revenue, We tested the regression analysis for (AVG_CLICKS_PER_VISIT, CROSS_SELL_SUCCESS, CANCELLATIONS_AFTER_NOON, MOBILE_LOGINS, PACKAGE_LOCKER, REFRIGERATED_LOCKER, AVG_CLICKS_PER_VISIT) with REVENUE, as those are the variables that most negatively affecting the APP revenue.
#### With our opservation to the second regression we can see a significant negative impact for "AVG_CLICKS_PER_VISIT" variable on the REVENUE with ZERO P Value and high negative Coeffeciency.
```

```python
#### IMPORT STATMODE ####
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
import statsmodels.formula.api 
lm=smf.ols(formula='REVENUE~AVG_CLICKS_PER_VISIT+CROSS_SELL_SUCCESS+CONTACTS_W_CUSTOMER_SERVICE+PRODUCT_CATEGORIES_VIEWED+AVG_TIME_PER_SITE_VISIT+CANCELLATIONS_BEFORE_NOON+CANCELLATIONS_AFTER_NOON+MOBILE_LOGINS+PC_LOGINS+WEEKLY_PLAN+EARLY_DELIVERIES+LATE_DELIVERIES+PACKAGE_LOCKER+REFRIGERATED_LOCKER+AVG_PREP_VID_TIME+MASTER_CLASSES_ATTENDED+MEDIAN_MEAL_RATING+AVG_CLICKS_PER_VISIT+TOTAL_PHOTOS_VIEWED', data=newacdata).fit()
print(lm.summary())
```

```python
lm=smf.ols(formula='REVENUE~AVG_CLICKS_PER_VISIT+CROSS_SELL_SUCCESS+CANCELLATIONS_AFTER_NOON+MOBILE_LOGINS+PACKAGE_LOCKER+REFRIGERATED_LOCKER+AVG_CLICKS_PER_VISIT', data=newacdata).fit()
print(lm.summary())
```
