```python


# fast way to define variables

exp_var = "CROSS_SELL_SUCCESS"

for i in new_acdata.columns:
    if i != "REVENUE" and i != "CROSS_SELL_SUCCESS": 
        exp_var = exp_var + " + " + i

# ols definition
sm_ols = sm.ols(formula = f"REVENUE ~ {exp_var}",
                  data    = new_acdata)


# fit the model
results = sm_ols.fit()


# print the results
print(results.summary())
```
