```python
# number of posts of each type
for tp in df["Type"].unique():
    ntp = len(df[df["Type"] == tp])
    print(f" Post type {tp} occcurs {ntp} times")
```

```python
# fraction of paid entries
df["Paid"].sum()/ df.shape[0]
```

```python
# average shares of ("share"-column)
#photo-type posts (.mean() in pandas
df[df["Type"] == 'Photo']["share"].mean()
```

```python
# print average likes of each post type
for tp in df["Type"].unique():
    like = df[df["Type"] == tp]["like"].mean()
    print(f" Post type {tp} was liked {int(like)} times")
```
