# Regression Analysis
**Conducting a regression analysis with the employee data.**

```python
# Import Statsmodel functions:
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
# run a bivariate regression with statsmodel
lm= smf.ols(formula='salary~jobcat',data=data).fit()
print(lm.summary())
```

```python
# run a bivariate regression with statsmodel
lm= smf.ols(formula='salary~jobcat+educ',data=data).fit()
print(lm.summary())
```

```python
# run a bivariate regression with statsmodel
lm= smf.ols(formula='salary~jobcat+salbegin+educ',data=data).fit()
print(lm.summary())
```

```python
# runing bivariate regression with the variables that effected the salary negatively
lm= smf.ols(formula='salary~jobtime+prevexp+minority+id',data=data).fit()
print(lm.summary())
```
